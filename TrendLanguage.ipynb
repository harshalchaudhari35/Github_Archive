```python
import pandas as pd

from plotly.offline import download_plotlyjs,plot,iplot,init_notebook_mode
init_notebook_mode(connected=True)
from plotly import tools

import plotly.graph_objs as go
```

```python
df = pd.read_csv('Language.csv')

df =df.truncate(after=381)
```

```python
trace0 = go.Scatter(x=list(df.Date),
                   y=list(df.Python), name='Python')

trace1 = go.Scatter(x=list(df.Date),
                   y=list(df.HTML), name='HTML')

trace2 = go.Scatter(x=list(df.Date),
                   y=list(df.Java), name='Java')

trace3 = go.Scatter(x=list(df.Date),
                   y=list(df.Ruby), name='Ruby')

trace4 = go.Scatter(x=list(df.Date),
                   y=list(df.JavaScript), name='JavaScript')
```

```python
data = [trace0, trace1, trace2, trace3, trace4]
layout = dict(
    title='Time series with range slider and selectors',
    xaxis=dict(
        rangeselector=dict(
            buttons=list([
                dict(count=1,
                     label='1m',
                     step='month',
                     stepmode='backward'),
                dict(count=6,
                     label='6m',
                     step='month',
                     stepmode='backward'),
                dict(count=1,
                    label='YTD',
                    step='year',
                    stepmode='todate'),
                dict(count=1,
                    label='1y',
                    step='year',
                    stepmode='backward'),
                dict(step='all')
            ])
        ),
        rangeslider=dict(
            visible = True
        ),
        type='date'
    )
)
```

```python
fig = dict(data=data, layout=layout)
iplot(fig)
```
